```r
library(ggplot2)
library(reshape2)
```

```r
yax2 = c(14.593,24.451,44.363,62.325,840.112,1920.011)
xax2 = c(1,5,10,20,50,70)
df2 <- data.frame(xax2,yax2)
```

## Inference from Plots - 

We compare two types of plots - Trigrams and Digrams in word coccurence and we derive the following inferences - 

1. For Digrams - Running time increases exponentialy with number of latin files (after 20 files) used but it is still scalable.

1. For Trigrams - Running time increases exponentialy with number of latin files (after 10 files). It can be noted that even for 30 files, the running time is very high. Thus Trigram word-coccurence doesn't have good scalabality when compared to digrams. This is mainly due to the fact that the digram algorithm is O(n2) and the trigram algorithm is O(n3).

Thus we infer that digram word cooccurence is more scalable than trigram word coccurence.

```r
plot(df2,type = 'o',col = "red",xlab = "",ylab = "")
title(main="Running Time in seconds - Digram vs Number of latin files", col.main="green", font.main=4)
title(xlab="Number of latin files", col.lab=rgb(0,0.5,0))
title(ylab="Running Time in seconds", col.lab=rgb(0,0.5,0))
```
